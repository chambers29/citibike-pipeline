```python
container_path = "abfss://citibike@databricksjm.dfs.core.windows.net"

df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(f"{container_path}/test")
)

display(df)
```

```python
df.printSchema()
```

```python
df.count()
```

```python
from pyspark.sql.functions import col, min, max
display(
    df
    .select(
        min(col("started_at")),
        max(col("ended_at")),
    )
)
```

```python
display(
    df
    .select(
        col("rideable_type")
    )
    .distinct()
)
```

```python
display(
    df
    .select(
        col("member_casual")
    )
    .distinct()
)
```

```python
df_edit = (
    df
    .withColumn(
        "duration_sec",
        col("ended_at").cast("long") - col("started_at").cast("long")
    )
)
```

```python
from pyspark.sql.functions import col, min, max, avg

display(
    df_edit
    .agg(
        min(col("duration_sec")),
        max(col("duration_sec")),
        avg(col("duration_sec"))
    )
)
```

```python
display(
    df_edit
    .filter(
        col("duration_sec") == 4578148
    )
)
```

```python
display(
    df_edit
    .sort(col("started_at"))
)
```

```python
display(
    df_edit
    .sort(col("duration_sec"), ascending=False)
)
```

```python
df_edit2 = (
    df_edit
    .filter(col("duration_sec") < 89970)
)
display(df_edit2.sort(col("duration_sec"), ascending=False).limit(10))
```

```python
df.count() - df_edit2.count()
```

```python
from pyspark.sql.functions import countDistinct, col, count

display(
    df_edit2
    .select(
        countDistinct(col("ride_id")),
        count(col("ride_id"))
    )
)
```
